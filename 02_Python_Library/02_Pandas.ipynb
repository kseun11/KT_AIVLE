```python
# 범위 지정
data.loc[data['Age'].between(25, 30)]
# data.loc[(data['Age'] >= 25) & (data['Age'] <= 30)]
```

```python
# Series
data.loc[data['MonthlyIncome'] >= 10000, 'Age']
```

```python
# 조건에 맞는 여러 열 조회
data.loc[data['MonthlyIncome'] >= 10000, ['Age', 'MaritalStatus', 'TotalWorkingYears']]
```

## 4. 데이터프레임 집계

```python
# 데이터 읽어오기
path = 'https://raw.githubusercontent.com/DA4BAM/dataset/master/Attrition_simple2.CSV'
data = pd.read_csv(path)  

# 상위 5개 확인
data.head(5)
```

#### 1) 열 하나 집계

특정 열의 값 합

```python
data['MonthlyIncome'].sum()
```

```python
data[['MonthlyIncome', 'TotalWorkingYears']].mean()
```

groupby

- as_index=False) 행 번호를 기반으로 한 정수 값이 인덱스로 설정.

```python
# MaritalStatus 별 Age 평균 --> 시리즈
data.groupby('MaritalStatus', as_index = True)['Age'].mean()
```

```python
# MaritalStatus 별 Age 평균 --> 데이터프레임
data.groupby('MaritalStatus', as_index = True)[['Age']].mean()
```

#### 2) 여러 열 집계

```python
data.groupby('MaritalStatus', as_index = False)[['Age', 'MonthlyIncome']].mean()
```

```python
# 기준열 이외의 모든 열에 대한 집계
data.groupby('MaritalStatus', as_index = False).sum()
```

```python
# 기준이 여러 열일 때
data.groupby(by = ['MaritalStatus', 'Gender'], as_index=False)[['Age','MonthlyIncome']].mean()
```

```python
# 정렬하기
data_sum = data.groupby(by = ['MaritalStatus', 'Gender'], as_index=False)[['Age','MonthlyIncome']].mean()
data_sum.sort_values(by = 'Age', ascending = True)
```

#### 3) 여러 함수로 한꺼번에 집계

```python
data.groupby('MaritalStatus', as_index = False)[['MonthlyIncome']].agg(['min', 'max', 'mean'])
```

## 6. 데이터프레임 변경

```python
import pandas as pd
import numpy as np
```

```python
# 데이터 읽어오기
path = 'https://raw.githubusercontent.com/DA4BAM/dataset/master/Attrition_simple2.CSV'
data = pd.read_csv(path)  

# 상위 5개 확인
data.head(5)
```

#### 1) 열 이름 변경

```python
# 일부 열 이름
data.rename(columns={'DistanceFromHome' : 'Distance', 
                    'EmployeeNumber' : 'EmpNo',
                    'JobSatisfaction' : 'JobSat',
                    'MonthlyIncome' : 'M_Income',
                    'PercentSalaryHike' : 'PctSalHike',
                    'TotalWorkingYears' : 'TotWY'}, inplace=True)

# 확인
data.head()
```

```python
# 모든 열 이름
data.columns = ['Attr','Age','Dist','EmpNo','Gen','JobSat','Marital','M_Income', 'OT', 'PctSalHike', 'TotWY']

# 확인
data.head()
```

#### 2) 열 추가

```python
# final_amt 열 추가
data['Income_LY'] = data['M_Income'] / (1+data['PctSalHike']/100 )
data['Income_LY'] = round(data['Income_LY'])
# 확인
data.head()
```

```python
# 풀이
data['JobSat2'] = np.where(data['JobSat'].isin([1, 2]), 'unsat', 'sat')
data['Diff_Income'] = data['M_Income'] - data['Income_LY']
data.head()
```

#### 3) 열 삭제

```python
# 열 하나 삭제
data.drop('Income_LY', axis = 1, inplace = True)
data.head()
```

```python
# 여러 열 삭제
data.drop(['JobSat2','Diff_Income'], axis=1, inplace=True)

# 확인
data.head()
```

#### 4) 유익한 기능

- map()
    - 범주형 값을 다른 값으로 쉽게 변경.

```python
# map()
data['Gen'] = data['Gen'].map({'Male': 1, 'Female': 0})

data.head()
```

- cut()
    - 크기를 기준으로 범위를 나누어 등급을 지정.
    - (] : a < x <= b

```python
pd.cut(data['M_Income'], 3)
```

```python
data['M_Income_Group'] = pd.cut(data['M_Income'], 3, labels = ['a', 'b', 'c'])
data.head()
```

```python
# 범위 확인
data.groupby('M_Income_Group', as_index = False)['M_Income'].agg(['min', 'max'])
```

```python
# 등급 구하기
bin = [0, 10000, 15000, np.inf]
data['M_Income_Group2'] = pd.cut(data['M_Income'], bins=bin, labels=['a', 'b', 'c'])

# 범위 확인
data.groupby('M_Income_Group2', as_index = False)['M_Income'].agg(['min', 'max'])
```
