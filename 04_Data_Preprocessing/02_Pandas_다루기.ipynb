## Dataframe 다루기

### 1. Join (merge)

- 결합은 여러개의 데이터셋을 하나로 모으는 작업.

- Pandas에서 join은 굉장히 간단
- 자동으로 Key를 잡아줌.
- default는 inner join.
    - how에 값을 지정하여 outer join을 수행.
    - 'left', 'right', 'outer'

```python
sales = pd.read_csv("https://raw.githubusercontent.com/DA4BAM/dataset/master/sales.csv")
products = pd.read_csv("https://raw.githubusercontent.com/DA4BAM/dataset/master/products.csv")
```

```python
sales.head()
```

```python
sales.isna().sum()
```

```python
products.head()
```

```python
products.isna().sum()
```

```python
# ProductID가 Key
pd.merge(sales, products).head() # 공통된 칼럼이 있으면 그 칼럼을 key로 사용
```

```python
pd.merge(sales, products, how='left').head()
```

```python
pd.merge(sales, products, how='right').head()
```

```python
pd.merge(sales, products, how='outer').head()
```

### 2. Join (merge) + Groupby

```python
# sales와 products를 ProductID 기준으로 merge하여 새로운 데이터프레임으로 저장
data = pd.merge(sales, products, on = 'ProductID')
data.head()
```

```python
# Category별 총 매출액 (amt)
data.groupby(by = 'Category', as_index = False)['Amt'].sum()
```

```python
data2 = data.groupby(by = 'ProductName', as_index = False)['Amt'].sum()
data2.sort_values(by = 'Amt', ascending = False).head(10)
```

### 3. Concat

```python
data2 = data.groupby('Category', as_index=False)['Amt'].sum()
data3 = data.groupby('Category', as_index=False)['Qty'].sum()
```

```python
# 위 아래로 붙이기
pd.concat([data2, data3])
```

```python
# 옆으로 붙이기
pd.concat([data2, data3], axis = 1)
```

### 3. Rolling & Shift

- Rolling은 주로 시계열 데이터에 대한 이동평균값을 구할때, 행을 이동 시킬때 사용.

```python
stock = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/SK.csv') 
stock.head()
```

```python
# Data, Close, Volume만을 남기기.
stock = stock[['Date', 'Close', 'Volume']]
stock.head()
```

```python
# 기준일 포함 과거 3일의 평균
stock['Close_MA_3'] = stock['Close'].rolling(3).mean()
stock.head(10)
```

```python
# shift()안의 숫자를 변경하기
stock['Close_MA_3_lag1'] = stock['Close_MA_3'].shift()
stock.head(10)
```

```python
# rolling (이동평균)과 shift를 한꺼번에 하기.
stock['Close_MA_3_lag1_2'] = stock['Close'].rolling(3).mean().shift()
stock.head(10)
```

```python
# min_periods
stock['Close_MA_3(2)'] = stock['Close'].rolling(3, min_periods = 1).mean()
stock.head()
```

```python
stock['Close_MA_3(2)'] = stock['Close'].rolling(3, min_periods=2).mean()
stock.head()
```
