```python
!cd yolov3; python detect.py \
    --weights '/content/yolov3/pretrained/yolov3.pt' \
    --source '/content/yolov3/data/videos/' \
    --project '/content/yolov3/detected' \
    --name 'videos' \
    --img 640 \
    --conf-thres 0.5 \
    --iou-thres 0.4 \
    --line-thickness 2 \
    --exist-ok \
    --device 0
```
