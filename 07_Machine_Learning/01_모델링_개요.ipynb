## 무작정 따라하기

### 0. 환경준비

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings(action='ignore')
```

### 1. 데이터 로딩

```python
path = 'https://raw.githubusercontent.com/DA4BAM/dataset/master/advertising.csv'
data = pd.read_csv(path)

data.head()
```

- TV, Radio, Newspaper 광고비용 지출에 따른 판매액 데이터
- 광고비용에 따른 매출액

### 2. 데이터 전처리

#### 1) 데이터를 X와 y로 나누기
- X: features, input
- y: target, output

```python
target = 'Sales'
x = data.drop(target, axis=1)
y = data.loc[:, target]
```

```python
x.head()
```

```python
y.head()
```

#### 2) train, validation, test로 분할

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

### 3. 학습

```python
# 필요한 함수들 불러오기
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_absolute_error
```

```python
# 모델 선언하기
model = LinearRegression()
```

```python
# 학습
model.fit(x_train, y_train)
```

### 4. 예측
- 학습의 결과 모델이 만들어짐
- 모델이 얼마나 정확한지 검증

```python
pred = model.predict(x_test)
pred
```

### 5. 평가

```python
# 평균 오차
mean_absolute_error(y_test, pred)
```

## 실습1 : Diamond Price 예측

### 1. 데이터 로딩

```python
path = 'https://raw.githubusercontent.com/DA4BAM/dataset/master/diamonds.csv'
data = pd.read_csv(path)
data = data.sample(2000, random_state=_state=2022)
data.head()
```

### 2. 데이터 전처리

```python
# x, y 분할
target = 'price'
x = data.drop(target, axis=1)
y = data.loc[:, target]
```

```python
# 가변수화
col_cat = ['cut', 'color', 'clarity']
x = pd.get_dummies(x, columns=col_cat, drop_first=True)
```

```python
# train, test 분할
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

### 3. 학습

```python
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_absolute_error
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

### 4. 예측

```python
pred = model.predict(x_test)
pred
```

### 5. 평가

```python
mean_absolute_error(y_test, pred)
```
